### Problem 4.2

```python
from scipy.optimize import fsolve
import numpy as np
from functools import partial
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(mu, m, sig2):
    return np.log(mu)-0.5*np.log(sig2/mu**2+1)-m
```

```python
for k in [0,0.25,0.5,0.75,1,1.25,1.5,1.75,2]:
    mus=[i/100 for i in range(500,1000)]
    gs=[partial(f,i,k) for i in mus]
    sig2s=[fsolve(g,2) for g in gs]
    sigs=[np.sqrt(s) for s in sig2s]
    plt.plot(sigs,mus,label="m")
    plt.show()
```
